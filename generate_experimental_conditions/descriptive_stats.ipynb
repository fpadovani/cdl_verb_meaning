```python
import pandas as pd 
import re
import random

token_pattern = re.compile(r"\b\w+(?:'\w+)?\b|[^\w\s']")
```

```python
def tokenize(text):
    # Apply your regex
    tokens = token_pattern.findall(text)
    # Keep only alphabetic/alphanumeric tokens (drop punctuation tokens)
    tokens = [t for t in tokens if re.match(r"^\w+(?:'\w+)?$", t)]
    return tokens
```

```python
def type_token_ratio(text):
    tokens = tokenize(text)
    if not tokens:
        return 0
    types = set(tokens)
    return len(types) / len(tokens)
```

```python
def bigram_ratio(text):
    tokens = tokenize(text)
    if len(tokens) < 2:
        return 0
    bigrams = [(tokens[i], tokens[i+1]) for i in range(len(tokens)-1)]
    return len(set(bigrams)) / len(bigrams)
```

```python
def trigram_ratio(text):
    tokens = tokenize(text)
    if len(tokens) < 3:
        return 0
    trigrams = [(tokens[i], tokens[i+1], tokens[i+2]) 
                for i in range(len(tokens)-2)]
    return len(set(trigrams)) / len(trigrams)
```

```python
def lexical_metrics_avg(dataset, sample_ratio=0.1, n_samples=5, seed=42):
    """
    dataset: list of strings (sentences)
    sample_ratio: fraction of sentences to sample each run
    n_samples: number of random sampling runs
    Returns: dict with average TTR, bigram ratio, trigram ratio
    """
    random.seed(seed)
    ttr_list, bigram_list, trigram_list = [], [], []
    n_sentences = len(dataset)
    sample_size = max(1, int(n_sentences * sample_ratio))
    
    for _ in range(n_samples):
        sample_sentences = random.sample(dataset, sample_size)
        # Concatenate all sampled sentences into one text
        text = " ".join(sample_sentences)
        tokens = tokenize(text)
        
        ttr_list.append(type_token_ratio(tokens))
        bigram_list.append(bigram_ratio(tokens))
        trigram_list.append(trigram_ratio(tokens))
    
    avg_metrics = {
        "TTR_avg": sum(ttr_list)/n_samples,
        "BigramRatio_avg": sum(bigram_list)/n_samples,
        "TrigramRatio_avg": sum(trigram_list)/n_samples
    }
    return avg_metrics
```

```python
def split_sentences(text):
    # Keep ., !, ? as sentence boundaries
    sentences = re.split(r'(?<=[.!?])\s+', text)
    # Remove any empty strings
    return [s.strip() for s in sentences if s.strip()]

# Tokenize a sentence
def tokenize(sentence):
    tokens = token_pattern.findall(sentence)
    tokens = [t for t in tokens if re.match(r"^\w+(?:'\w+)?$", t)]
    return tokens

# Compute average sentence length
def average_sentence_length(text):
    sentences = split_sentences(text)
    if not sentences:
        return 0
    sentence_lengths = [len(tokenize(s)) for s in sentences]
    return sum(sentence_lengths) / len(sentence_lengths)
```

## Wikipedia

```python
with open('./corpora/WIKI_rand/original/wiki_rand.train.txt', 'r') as f:
    dataset = f.read()
    dataset = dataset.lower()
    
print("TTR =", type_token_ratio(dataset))
print("Bigram Ratio =", bigram_ratio(dataset))
print("Trigram Ratio =", trigram_ratio(dataset))
average_sentence_length(dataset)
```
